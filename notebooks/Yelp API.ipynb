```python
def do_yelp_request(restaurants_df):
    is_open_list = []
    not_found_ids_list = []
    
    for index, row in restaurants_df.iterrows():
        name = row['name']
        
        url = URL + index
        
        request = requests.get(url, headers = HEADERS)
    
        status_code = request.status_code
        
        if status_code != 200:
            print "Error status code (%d) for name: %s"%(status_code, name)
        else:
            response = json.loads(request.text)
            if name == response['name']:
                is_open = not response['is_closed']
                try:
                    price = response['price']
                except:
                    print "No price for %s. Setting it to 0"%(name)
                    price = '0'
                try:
                    rating = response['rating']
                except:
                    print "No rating for %s. Setting it to 0.0"%(name)
                    rating = '0.0'
                try:
                    review_count = response['review_count']
                except:
                    print "No review count for %s. Setting it to 0"%(name)
                    review_count = '0'
                try:
                    is_claimed = response['is_claimed']
                except:
                    print "No is_claimed for %s. Setting it to False"%(name)
                    is_claimed = False

                try:
                    has_photos = len(response['photos']) > 0
                except:
                    print "No photos. Setting it to 0"%s(name)
                    has_photos = False

                coords = response['coordinates']
                latitude = coords['latitude']
                longitude = coords['longitude']

                #print "Found restaurant (%s) with Yelp API. Is open: %d"%(name, is_open)
                is_open_list.append( (index, name, is_open, price, rating, review_count, \
                                      latitude, longitude, is_claimed, has_photos) )       
    return is_open_list
```
